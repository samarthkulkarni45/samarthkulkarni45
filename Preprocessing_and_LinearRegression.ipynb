# Ridge-Lasso Regression

```python
# first standardize data
# imort preprocessing
from sklearn import preprocessing
```

```python
# we creste scaler object which stores scaling info
scaler=preprocessing.StandardScaler().fit(x_train)
```

```python
#transform xtrain into scaler
x_train_s=scaler.transform(x_train)
```

```python
x_test_s=scaler.transform(x_test)
```

```python
# we have standardized
# ridge rigresson

from sklearn.linear_model import Ridge
```

```python
lm_r=Ridge(alpha=0.5) #lambda here in python is alpha
lm_r.fit(x_train_s,y_train)
```

```python
# find r2 val
```

```python
r2_score(y_test,lm_r.predict(x_test_s,))
```

```python
# we can change alpha pr lamda and we want to find valu of alpha for which or r2 val is max
# we use validation curv
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
param_range=np.logspace(-2,8,100)
# logspace will create 100 values from 10 raised to pow -2 and 10 raised to power 8
```

```python
param_range
```

```python
train_scores,test_scores=validation_curve(Ridge(),x_train_s,y_train,"alpha",param_range,scoring="r2")
```

```python
# we have train this model for all lambda/alpha values
#notice we have not given our "test set" to validation curve
# so how this validation curve calculating test curve?
# this is actually running a K-fold validation on our data
print(train_scores)
print(test_scores)
#each value of lambda/alpha we have 3 values for r2
```

```python
# we will find mean of this
train_mean=np.mean(train_scores,axis=1)
```

```python
test_mean=np.mean(test_scores,axis=1)
```

```python
print(train_mean) #for each of our lambda val these following r2 values that we are getting
```

```python
max(test_mean)
```

```python
# here we have improved our r2 val from 0.5 to 0.7--
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean)  #lamb ar in expo form we take log ot thrm
```

```python
#as we are increasing our lambda our r2 value slightyly increasig and then decresing and then it around 0 for very hifg lambda
# next we will find lambda val for which r2 is max
# to find its location
np.where(test_mean==max(test_mean))
```

```python
#at 31st index

param_range[31]
```

```python
#now we will fit our Ridge model for this lambda value
lm_r_best=Ridge(alpha=param_range[31])
```

```python
#lets fir


lm_r_best.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lm_r_best.predict(x_test_s))
```

```python
r2_score(y_train,lm_r_best.predict(x_train_s))
```

```python
#we will not see much diffrent in values from Linear Regresson and Ridge Regression as we have small data 
```
